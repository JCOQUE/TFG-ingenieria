```python
import pandas as pd
import xgboost as xgb
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV

from tfg_module import my_get_time_series as mgts
from tfg_module import my_process_data as mpd
from tfg_module import my_future as mf
```

```python
model_name = 'XGBoost'
TARGET = 'Compras' # target can be either Ventas or Compras
```

```python
ts = mgts.get_ts(TARGET)
ts
```

```python
X, y = mpd.create_features(ts, target = TARGET, informer = False)
```

```python
cross_val_split = TimeSeriesSplit(n_splits=3, test_size=20)
model_XGB = xgb.XGBRegressor()
params_grid = {'max_depth': [3,5,10],
                        'n_estimators': [50, 500, 1000]#,
                        # 'tree_method':['exact', 'approx']
                        # 'learning_rate': [0.1, 0.01],
                        # 'n_estimators': [50, 500, 1000],
                        # 'colsample_bytree': [0.3,  0.7],
                        # 'max_depth': [3, 5, 10],
                        # 'max_leaves': [1,2,3,4,5]
                    }

# SEE EARLY STOPPING ROUNDS AND CALLBACKS https://xgboost.readthedocs.io/en/stable/python/python_api.html
metrics_to_keep = ['neg_mean_absolute_error', 'neg_root_mean_squared_error']
best_model_XGB = GridSearchCV(estimator=model_XGB, cv=cross_val_split, param_grid=params_grid,
                         scoring = metrics_to_keep,
                        refit = 'neg_mean_absolute_error', return_train_score = True)
```

```python
best_model_XGB.fit(X, y)
```

```python
results = best_model_XGB.cv_results_
```

```python
opposite_metric = {'neg_mean_absolute_error':'neg_root_mean_squared_error',
                   'neg_root_mean_squared_error':'neg_mean_absolute_error'}
```

```python
models_by_metric = {}
for metric in metrics_to_keep:
    best_index = results[f'rank_test_{metric}'].argmin()
    best_score = results[f'mean_test_{metric}'][best_index]*-1
    other_metric_score = results[f'mean_test_{opposite_metric[metric]}'][best_index]*-1
    best_params = results['params'][best_index]
    best_model = xgb.XGBRegressor(**best_params)
    best_model.fit(X,y)
    if metric == 'neg_mean_absolute_error':
        models_by_metric[metric] = (best_model, best_params, best_score, other_metric_score)
    else:
        models_by_metric[metric] = (best_model, best_params, other_metric_score, best_score)

best_xgb_models_df = pd.DataFrame(models_by_metric)
best_xgb_models_df.rename(columns = {'neg_mean_absolute_error':'best_MAE', 'neg_root_mean_squared_error':'best_RMSE'}, inplace = True)
best_xgb_models_df.index = ['model', 'parameters', 'mae', 'rmse']
```

```python
best_xgb_models_df
```

```python
best_mae_model = best_xgb_models_df.loc['model','best_MAE']
best_mae_model
```

```python
best_mae_df = mf.get_pred_df(ts, best_mae_model)
best_mae_df
```

```python
mf.save_pred_plot(model_name, ts, best_mae_df, 'mae')
```

```python
best_rmse_model = best_xgb_models_df.loc['model','best_RMSE']
best_rmse_model
```

```python
best_rmse_df = mf.get_pred_df(ts, best_rmse_model)
best_rmse_df
```

```python
mf.save_pred_plot(model_name, ts, best_rmse_df, 'rmse')
```
