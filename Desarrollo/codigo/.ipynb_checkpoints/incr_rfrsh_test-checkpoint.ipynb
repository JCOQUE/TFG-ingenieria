```python
import pandas as pd
from datetime import datetime
```

```python
def format_with_commas(x):
    return '{:,.0f}'.format(x)


def add_grupo(asiento):
    grupo = asiento['NoCuenta'][:1]
    
    return grupos_plan_general_contable[grupo]


def create_purchases(asiento):
    if asiento['NoCuenta'].startswith('6'):
        return asiento['Debe'] - asiento['Haber']
    else:
        return 0
    
    
def create_sales(asiento):
    if asiento['NoCuenta'].startswith('7'):
        return asiento['Haber'] - asiento['Debe']
    else:
        return 0
```

```python
pd.options.display.float_format = format_with_commas
```

```python
dataset = pd.read_csv('../datasets/diario.csv')
dataset.head(6)
```

```python
dataset.drop(columns = ['Movimiento'], inplace = True)
dataset
```

```python
dataset.reset_index(inplace = True)
dataset.rename(columns = {'index': 'ID'}, inplace = True)
dataset
```

```python
dataset['NoCuenta'] = dataset['NoCuenta'].astype(str).str[:3]
dataset['Compras'] = dataset.apply(create_purchases, axis = 1)
dataset['Ventas'] = dataset.apply(create_sales, axis = 1)
dataset
```

```python
grupos_plan_general_contable = {'1': '100',
                               '2':'200',
                               '3': '300',
                               '4':'400',
                               '5': '500',
                               '6':'600',
                               '7': '700',
                               '8':'800',
                               '9': '900'}
dataset['NoGrupo'] = dataset.apply(add_grupo, axis =1)
dataset
```

```python
dataset['Fecha'] = pd.to_datetime(dataset['Fecha']).dt.strftime('%d/%m/%Y')
dataset
```

```python
dataset_incr = dataset.copy()
```

```python
dataset_incr['Fecha'] = pd.to_datetime(dataset_incr['Fecha'], format='%d/%m/%Y')
dataset_incr
```

```python
dataset_incr = dataset_incr[~((dataset_incr['Fecha'].dt.year == 2017) & (dataset_incr['NoCuenta'].str.startswith('7')))]
```

```python
dataset_incr.head(30)
```

```python
for _ in range(200):
    dataset_incr = pd.concat([dataset_incr, dataset_incr.tail(100)], ignore_index=True)
```

```python
dataset_incr.shape
```

```python
dataset_incr['Fecha'] = dataset_incr['Fecha'].dt.strftime('%d/%m/%Y')
dataset_incr
```

```python
dataset_incr.drop(columns = 'ID', inplace = True)
dataset_incr.reset_index(inplace = True)
dataset_incr.rename(columns = {'index': 'ID'}, inplace = True)
dataset_incr
```

```python
dataset_incr.to_parquet('../datasets/diario.parquet')
dataset_incr.to_csv('../datasets/check.csv')
```
